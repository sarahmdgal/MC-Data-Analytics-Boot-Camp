```python
stats_df = df.drop(['Unnamed: 0', 'x', 'y', 'z'], axis=1)
stats_df.head()
```

```python
corr = stats_df.corr()

sns.heatmap(corr, vmin=-1, annot=True)
```

```python
#create color palette
#color Hex numbers are from Bulbapedia (https://bulbapedia.bulbagarden.net/wiki/Category:Type_color_templates)
diamond_type_colors=[
                    '#A8B820',  # carat
                    '#A8A878',  # depth
                    '#A040A0',  # table
                    '#F8D030',  # price
                   ]
```

```python
plt.subplots(figsize=(13,7))
sns.violinplot(x='color', y='price', data=stats_df, palette=diamond_type_colors)
```

```python
plt.subplots(figsize=(10,6))
sns.swarmplot(x='color', y='price', data=stats_df, palette=diamond_type_colors)
```

```python
#combine a violin plot with a swarm plot
plt.figure(figsize=(13,7))
sns.violinplot(x='color',
               y='price',
               data=stats_df,
               inner=None, #removes bars inside violin plot
               palette=diamond_type_colors)

sns.swarmplot(x='color',
              y='price',
              data=stats_df,
              color='k', #make points black
              alpha=0.7, #slightly transparent
              palette=diamond_type_colors)

plt.title('Color By Price')
```
